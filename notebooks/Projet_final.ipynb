```python
import cv2
import collections
import matplotlib.image as image

df_ok = pd.DataFrame()

df_not_ok = pd.DataFrame()
```

```python
indice = 0

for imag in file_list_test_ok:
    img=cv2.imread(imag)
    rows = []
    for i in range(512):
        for j in range(512):
            rows.append(img[i][j][0])

    number = dict(collections.Counter(rows))
    
    indice += 1
    df_ok = pd.concat([df_ok,pd.DataFrame(number,index=[indice])])

```

```python
df_ok["Target"] = 1
df_ok
```

```python
indice = 0

for imag in file_list_test_not_ok:
    img=cv2.imread(imag)
    rows = []
    for i in range(512):
        for j in range(512):
            rows.append(img[i][j][0])

    number = dict(collections.Counter(rows))
    
    indice += 1
    df_not_ok = pd.concat([df_not_ok,pd.DataFrame(number,index=[indice])])
```

```python
df_not_ok["Target"] = 0
df_not_ok
```

```python
df_image = df_ok.append(df_not_ok, ignore_index=True)
```

```python
df_image.fillna(0,inplace=True)
df_image
```

```python
from sklearn.utils import shuffle
df_image = shuffle(df_image)
df_image
```

```python
df_image.to_csv("df_images.csv",index=False)
```
